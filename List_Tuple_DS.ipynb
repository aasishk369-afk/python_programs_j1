# List in data strutures 
1. list is collection of elements enclosed by []
2. list a mutable - editable 
3. it's allowed duplicate elements
4. each element seperated by ,
   

```python
# create list
names=["aasish","eswar","taman","TDP<3<3"]
ages=[45,54,5,55,65,88]
print(names)
print(ages)
print(type(names))
print(type(ages))
```

```python
#concatenation of list
a=[10,20,30,40,50]
b=[40,50,60,70,80]
result=a+b
print(result)
```

```python
#repeating list
ages=[10,20,30]
print(ages*69)
```

```python
# accesing elements from list - index
ages=[10,20,30,40,50,60,70]
print(ages)
print(ages[4])
print(ages[-3])
```

```python
#methods or funtion in list
print(dir(list))
```

```python
#description
print(help(list))
```

```python
#add elements to list append(),extend(),insert()
a=[10,20,30,40,50]
print(a)
a.append(100)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.extend([100,200,300])
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.insert(1,100)
print(a)
```

```python
#remove elements from list -pop(),remove(),clear()
a=[10,20,30,40,50]
print(a)
a.pop(4)
```

```python
a=[10,20,30,40,50]
print(a)
a.remove(50)
print(a)
```

```python
# sorting techniques - sort(reverse=True or reverse=False)
a=[6,4,5,3,1,2]
print(a)
a.sort(reverse=True) # True means Descending , False means Ascending
print(a)
```

# tuples in data strutures
1. tuple is collection of elements enclosed by ()
2. tuple is immutable - not editable
3. each element seperated ,
4. duplicates allowed

```python
a=(10,20,30,40,50)
names=("aasish","aryan",10.25,12,56)
print(a)
print(names)
print(type(a))
print(type(names))
```

```python
#concatenation,repeating, Accesing elements from tuple 
```

```python
# concatenation 
t1 = (9941, 98, 5)
t2 = (656962, 554)

result = t1 + t2
print(result)   
```

```python
#repeating

WORD = ('ACETIC', 'BROCKK')

result = WORD * 24
print(result) 
```

```python
#Accesing elements from tuple 

NUM = (56, 200,874,478)

print(NUM[1])  
print(NUM[2])   
```

```python
#methods
print(dir(tuple))
```

```python
# count()
a=(10,10,20,30,10,40)
print(a.count(10))
```

```python
# count()
a=(10,10,20,30,10,40)
print(a.count(100))
```

```python
#index()
a=(10,20,30,40,50)
print(a.index(40))
```
