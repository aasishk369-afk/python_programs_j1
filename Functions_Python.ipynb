````python
#write a program to print sum of two numbers using function
def add_number():
    a=int(input("enter first number"))
    b=int(input("enter second number"))
    result=a+b
    print(result)
    return result
add_number()
````

````python
#write a program to print addition of two string using function 
def add_string():
    a=str(input("enter first string"))
    b=str(input("enter second string"))
    result=a+ b
    print(result)
    return result
add_string()
````
