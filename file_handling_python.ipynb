# file handling in python 
1. file is collection of information and stored in memory
2. files are 2 types `1.text files` and `2.binary files`
3. text files are `.txt, .py, .doc, .html, .ipynb ....etc`
4. binary files - executions files ` .exe`

# file Operations 
1. create file - w mode
2. write file - w mode
3. read file - r mode
4. append file - a mode

```python
# write a program to create file usig python script
create_file=open("profile.txt","w")
print("file created successfully and verified in home directory")
create_file.close()
```

```python
# write a program to write info to profile.txt using python script
write_file=open("profile.txt","w")
data="welcome to python program \nthis program to write info to file"
write_file.write(data)
print("successfully info write to file please verify the file")
write_file.close()
```
