````python
#write a program a iterate characters from string using for loop?
name="Aasish"
for char in name:
    print(char)
````

````python
#write aprogram to print even numbers in between 500 to 600 using for loop?
for n in range (500,601):
    if (n%2 == 0):
        print(n,end=",")
    
````

````python
#write a program to print odd numbers in between 100 to 200 using for loop?
for n in range (100,200):
    if (n%2 == 1):
        print(n,end=",")
````

````python
#write aprogram to find the vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char in "aeiouAEIOU"):
        print(char)
````

````python
#write a program to find the nonvowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if(char not in "aeiouAEIOU"):
        print(char)
````
