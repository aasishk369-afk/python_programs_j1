# dict 
1. dict is collection of items enclosed by {} - item is a key:value pair

```python
info={"name":"aasish","salary":10000.25,"address":"hyderabad","age":25}
print(info)
print(type(info))
```

```python
# read data and update data from dict
info={"name":"shiva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
print(info["address"])
```

```python
info={"name":"aasish","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info["address"]="banglore"
print(info)
```

```python
#methods or functions in dict

print(dir(dict))
```

```python
info={"name":"aasish","salary":10000.25,"address":"hyderabad","age":25}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
#update()
info={"name":"aasish","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info.update({"salary":100000.50})
print(info)
```

```python
#pop()
info={"name":"shiva","salary":10000.25,"address":"hyderabad","age":25}
print(info)
info.pop("salary")
print(info)
```

# sets
1. set is a collection of unorder unique elements enclosedd by {}
   


```python
# create set
a={2,4,1,6,4,5,8,7,}
print(a)
print(type(a))
```

```python
#methods or functions 
print(dir(set))
```

```python
#add 
a={20,30,40}
print(a)
a.add(100)
print(a)
```

```python
#union
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.union(b)
print(c)
```

```python
#intersection 
a={1,2,3,4,5,6}
b={4,5,6,7,8,9,}
c=a.intersection(b)
print(c)
```

```python
#discard
a={1,2,3,4,5,6}
print(a)
a.discard(6)
print(a)
```
