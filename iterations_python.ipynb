# Iterations  in python
1. Iterations are loops to use work on the repetative tasks or repetative performences
2. Generally iterations are of 2 types
3. 'for' loop
4. while loop

```python - for loop
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter_variable is a variable to iterate the data from sequence")
    print("in is the keyword to validate the sequence is present or not")
    print("sequence is collection like range,string,list,tuple....etc")
```

```python - while loop
while (condition):
    print("while is the keyword to start the loop and validate conditions based on values")
    print("If condition is true then inside statements are executed")
```




 

````python
# Write a program to print 1 to 60 numbers using for loop.
for num in range(50,61):
    print(num)
````

````python
#Write a program to iterate characters from string using for loop.
name="Hemanth"
for char in name:
    print(char)
````

````python
#Write a program to print even numbers in between 500 to 600 using for loop.
for num in range(500,601):
    if(num%2==0):
        print(num,end=",")
````

````python
#Write a program to print odd numbers in between 500 to 600 using for loop.
for num in range(500,601):
    if(num%2==1):
        print(num,end=",")
````

````python
#Write a program to find the vowels from string using for loop 
name=str(input("PLease enter your name"))
for char in name:
    if(char in "AEIOUaeiou"):
        print(char)
````

````python
#Write a program to find the non-vowels from string using for loop 
name=str(input("PLease enter your name"))
for char in name:
    if(char not in "AEIOUaeiou"):
        print(char)
````

````python
#write a program to iterate names from list of names
names=["shivaji","vikas","hemanth","aditya"]
for n in names:
    print("my name is:",n)
````
